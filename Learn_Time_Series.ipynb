```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Bitcoin.csv')
df.head()
```

Time Series Index

```python
df['Date']
```

```python
# Convet 'Date' To a datetime and set as index
pd.to_datetime(df['Date'], format = '%Y-%m-%d')
df.set_index('Date', inplace = True)
df.head()
```

```python
# Select the Bitcoin Data for November 2021
df.loc['2021-11-09']
```

```python
df1 = pd.read_csv('/content/Bitcoin.csv', parse_dates=True, index_col = ['Date'])
df1.index
```

```python
# Ensure the index is a DatetimeIndex
df.index = pd.to_datetime(df.index)
```

```python
# Resampling to Montly Frequency and calculate the mean closing
df.resample('M').mean()
```

```python
# Resampling to Weekly Frequency and calculate the mean closing
df.resample('w').mean()
```

## **Exploring Data**

```python
df['7_day_rolling']=df['Close'].rolling(window=7).mean()
df[['Close', '7_day_rolling']].loc['2021'].plot()
plt.show()
```

```python
# Find out the highest average month
df.resample('M').mean()['Close'].idxmax()
```

```python
# Calculate Daily Returns
df['daily returns'] = df['Close'].pct_change()*100
```

```python
# Days with more than 10% change in closing price
df[abs(df['daily returns'])>10]
```

### **Data Visualization**

```python
# Daily closing price
df['Close'].plot(title='Daily Closing price')
plt.show()
```

```python
# Plot the Yearly Volume
df.resample('Y').sum()['Volume'].plot(title='Yearly Volume')
plt.show()
```

```python
# Plotting Closing Price and a 30 Day rolling volume
df['3o_day_rolling_vol'] = df['Volume'].rolling(window=30).mean()
df['3o_day_rolling_vol'].plot(legend = True)
ax = df['Close'].plot(secondary_y=True, legend = True)
ax.set_ylabel('Volume')
plt.show()
```

```python
df['closed_lag1']=df['Close'].shift(1)
df['closed_lag2']=df['Close'].shift(2)
```
